```python
lista_aux = []
lista_filmes = []

while True:
    condicao = input("Digite quantos filmes deseja adicionar na sua lista ou 'Sair' para finalizar: ")
    if condicao.lower() == "sair":
        break
    else:
        condicao = int(condicao)
        i = 0
        while i < condicao:
            filmes = {}
            chave = input("Digite o nome do filme: ")
            if chave in lista_aux:
                print("Filme repetido!")
            else:
                while True:
                    valor = float(input("Digite a nota do filme (0 a 10): "))
                    if valor >= 0 and valor <=10:
                        break
                    else:
                        print("Nota inválida! Digite um valor entre 0 e 10.")
                filmes[chave] = valor
                lista_filmes.append(filmes)
                lista_aux.append(chave)
                i += 1

    print("Filmes salvos com sucesso!")

    painel = input("\nDigite:\n AD para Adicionar um Filme\n RM para Apagar um Filme\n ED para Editar um Filme\n ❌Sair❌ para Finalizar o Programa\n").lower()

    if painel == "ad":
        filmes = {}
        adicionar_filme = input("Digite o nome do filme que deseja inserir na lista: ")
        if adicionar_filme in lista_aux:
            print("Filme repetido!")
        else:
            while True:
                valor = float(input("Digite a nota do filme (0 a 10): "))
                if valor >= 0 and valor <=10:
                    break
                else:
                    print("Nota inválida! Digite um valor entre 0 e 10.")
            filmes[adicionar_filme] = valor
            lista_filmes.append(filmes)
            lista_aux.append(adicionar_filme)

    elif painel == "rm":
        remover_filme = input("Digite o nome do filme que deseja remover: ")
        if remover_filme in lista_aux:
            filme_encontrado = False
            for filme in lista_filmes:
                if remover_filme in filme:
                    lista_filmes.remove(filme)
                    lista_aux.remove(remover_filme)
                    filme_encontrado = True
                    break
            if filme_encontrado:
                print(f"Filme '{remover_filme}' removido com sucesso!")
            else:
                print("Ocorreu um erro ao tentar remover o filme.")
        else:
            print("Filme não encontrado!")

    elif painel == "ed":
        editar_filme = input("Digite o nome do filme que deseja editar: ")
        if editar_filme in lista_aux:
            while True:
                nova_nota = float(input("Digite a nova nota (0 a 10): "))
                if valor >= 0 and valor <=10:
                    break
                else:
                    print("Nota inválida! Digite um valor entre 0 e 10.")
            for filme in lista_filmes:
                if editar_filme in filme:
                    filme[editar_filme] = nova_nota
                    print(f"Nota do filme '{editar_filme}' atualizada com sucesso!")
        else:
            print("Filme não encontrado!")
    elif painel == "sair":
        break
media = 0
for filme in lista_filmes:
    for nome, nota in filme.items():
        media += nota

result_f = media / len(lista_filmes)

print(f"Média das notas: {result_f}")

print("\nLista de Filmes:")
print("-------------------------------")
for filme in lista_filmes:
    for nome, nota in filme.items():
        print(f"Filme: {nome} | Nota: {nota}")
print("-------------------------------")
```
